### Проверка корректности рассчёта эффективности обогащения
Корректность расчёта проверим с помощью метрики MAE (англ. Mean Absolute Error, «среднее абсолютное значение ошибки»). Вычислим эффективность обогащения на обучающей выборке для признака rougher.output.recovery. Найдём MAE между расчётным и измеренным значением признака, величины которого заданы в исходном наборе данных - признак rougher.output.recovery. Поскольку значения средней абсолютной ошибки всегда положительное вещественное число, стремящееся к нулю, то, чем ближе к нулевому значению будет расчётное, тем достовернее будет расчёт.

<p><center><b>Формула 1: Эффективность обогащения</b></center></p>

$$
Recovery = \frac{С * (F - T)}{F * (C - T)} * {100\%}
$$, где<br>
C — доля золота в концентрате после флотации/очистки (признак rougher.output.concentrate_au);<br>
F — доля золота в сырье/концентрате до флотации/очистки (признак rougher.input.feed_au);<br>
T — доля золота в отвальных хвостах после флотации/очистки (признак rougher.output.tail_au).

```python
# рассчётное значение эффективности обогащения
C = df_train['rougher.output.concentrate_au']
F = df_train['rougher.input.feed_au']
T = df_train['rougher.output.tail_au']
recovery_check= ((C * (F - T))/(F * (C - T)))* 100 

# фактическое значение эффективности обогащения
recovery_source = df_train['rougher.output.recovery']
```

```python
print('Средняя абсолютная ошибка между измеренными и расчётными значениями:\n',\
      mean_absolute_error(recovery_source, recovery_check))
```

Определим числовые характеристики распределения числовых рядов и выдвинем гипотезу о виде их распределения. В качестве гипотезы H0 принимаем утверждение, что выборка данных нормальное (Гауссовское) распределение; гипотеза H1 является противоположной гипотезе H0.

Данные тестового и обучающего наборов данных, определённые для анализа, проверим на нормальность, определив выборочные характеристики (среднее mean(), медиана meidan(), мода mode(), коэффициент асимметрии skew() и эксцесса kurtosis(); затем с помощью критериев Shapiro-Wilk и D'Agostino.

С помощью шаблона, представленного на Рисунке 1 можно визуально определить вид гистограммы распределения для каждого числового ряда.

histogram-examples.png

*Рисунок 1. Виды гистограмм распределения (источник: https://chartio.com/learn/charts/histogram-complete-guide/)*

Чтобы определить ассиметрию гистограммы, помимо визуального метода можно использовать числовой. В этом случае выполняется сравнение трех числовых параметров распределения: мода, медиана, среднее. В случае с левосторонней ассиметрией справедливо будет следующее отношение мода > медиана > среднее; для правосторонней среднее > медиана > мода. Однако, достаточно применения только двух параметров - среднее и медиана. Когда все три параметра равны - это означает симметричное распределение. Дополнительно, с помощью коэффициента ассиметрии A можно также определить скошенность гистограммы: при A < 0 гистограмма левосторонняя, при A > 0 - правосторонняя, A = 0 подтверждает симметричность гисторграммы. Если модуль значения коэффициента ассиметрии больше 0.5, то считается ассиметрия существенной, а при A < 0.25 - незначительной.

Для критериев Shapiro-Wilk и D'Agostino определим доверительную вероятность оценки (надёжность оценки) P при уровне значимости 𝛼 равным 0.05. Согласно формуле P = 1 − 𝛼, надёжность оценки будет равна 0.95. Эти величины справедливы для обоих критериев.

Таким образом, выбор между двумя гипотезами определяется сравнением расчётного значения коэффициента уровня значимости p-value и заранее принятого 𝛼 = 0.05. Отсюда, если p-value > 𝜶, то нет оснований отклонить
нулевую гипотезу – H0; если p-value ≤ 𝜶, то нулевая гипотеза отклоняется
и решение выносится в пользу альтернативы – H1.

Приступим к расчётам.

```python
# расчёт числовых характеристик
print('Расчёт числовых характеристик')

# расчёт для тестового набора
print('Тестовый набор данных')
print('Среднее арифметическое:', f'{np.mean(df_test["rougher.input.feed_size"]):3.1f}')
print('Медиана:', f'{np.median(df_test["rougher.input.feed_size"]):3.1f}')
print('Коэффициент ассиметрии:', f'{st.skew(df_test["rougher.input.feed_size"]):3.1f}')
print('Коэффициент эксцесса:', f'{st.kurtosis(df_test["rougher.input.feed_size"]):2.1f}')

# расчёт для обучающего набора
print('\nОбучающий набор данных')
print('Среднее арифметическое:', f'{np.mean(df_train["rougher.input.feed_size"]):3.1f}')
print('Медиана:', f'{np.median(df_train["rougher.input.feed_size"]):3.1f}')
print('Коэффициент ассиметрии:', f'{st.skew(df_train["rougher.input.feed_size"]):3.1f}')
print('Коэффициент эксцесса:', f'{st.kurtosis(df_train["rougher.input.feed_size"]):2.1f}')
```

Анализ распределений величин обучающего и тестового наборов данных подтвердил выводы визуального анализа. Т.о., можно достоверно утверждать, что эти два числовых ряда имеют положительную (правостороннюю) ассиметрию и островершинный эксцесс распределения.

Расчётные параметры числовых рядов различаются незначительно, однако, нельзя достоверно заявлять об идентичности распределений.

Принимая во внимание приведённую выше информацию, можно говорить о необходимости удаления выбросов методом наложения межквартильного диапазона (англ., Inter-Quartile Range, IQR).

```python
# удаление выбросов с помощью метода межквартильного диапазона (IQR)
# расчёт для тестового набора
Q1 = np.percentile(df_test['rougher.input.feed_size'], 25)
Q3 = np.percentile(df_test['rougher.input.feed_size'], 75)
IQR = Q3 - Q1
ul = Q3 + 1.5 * IQR
ll = Q1 - 1.5 * IQR
df_test = df_test.drop(df_test[df_test['rougher.input.feed_size'] < ll].index)
df_test = df_test.drop(df_test[df_test['rougher.input.feed_size'] > ul].index)

# расчёт для обучающего набора
Q1 = np.percentile(df_train['rougher.input.feed_size'], 25)
Q3 = np.percentile(df_train['rougher.input.feed_size'], 75)
IQR = Q3 - Q1
ul = Q3 + 1.5 * IQR
ll = Q1 - 1.5 * IQR
df_train = df_train.drop(df_train[df_train['rougher.input.feed_size'] < ll].index)
df_train = df_train.drop(df_train[df_train['rougher.input.feed_size'] > ul].index)
```

### Функция определения качества моделей машинного обучения - sMAPE

Качество будущих моделей машинного обучения будет определяться с помощью критерия кросс-валидации - sMAPE (англ., Symmetric Mean Absolute Percentage Error). Это критерий оценки качества в задачах классификациии регрессии и переводится как Симметричная  Средняя Абсолютная Процентная Ошибка. Этот коэффициент можно измерять в долях или процентах. Интерпретируется он следующим образом: если sMAPE = 11.4%, то это говорит о том, что ошибка составила 11,4% от фактических значений. Значения коэффициента всегда положительные. Лучшее значение 0.0. Напишем функцию расчёта sMAPE для её последующего применения.

<p><center><b>Формула 2: Симметричная  Средняя Абсолютная Процентная Ошибка (sMAPE)</b></center></p>

$$
sMAPE = \frac{1}{N}\sum_{i=1}^{N}\frac{|y_i - \hat{y_i}|}{(|y_i| + |\hat{y_i}|)/2}*100\%
$$

```python
# функция расчёта sMAPE
def smape(source, check):
    return 100/len(source) * np.sum(2 * np.abs(check - source) / (np.abs(source) + np.abs(check)))
```

Подготовим функцию к использованию в классе GridSearchCV().

```python
# для предотвращения появления предупреждения SettingWithCopy
pd.options.mode.chained_assignment = None

# список числовых признаков
numeric = features_rough_train.columns

# применяем метод StandardScaler для масштабирования признаков
scaler = StandardScaler()
scaler.fit(features_rough_train[numeric])
features_rough_train[numeric] = scaler.transform(features_rough_train[numeric])
features_rough_test[numeric] = scaler.transform(features_rough_test[numeric])

# список числовых признаков
numeric = features_fine_train.columns

# применяем метод StandardScaler для масштабирования признаков
scaler = StandardScaler()
scaler.fit(features_fine_train[numeric])
features_fine_train[numeric] = scaler.transform(features_fine_train[numeric])
features_fine_test[numeric] = scaler.transform(features_fine_test[numeric])

# выводим первые пять строк обучающей выборки
display(features_rough_train.head())
display(features_fine_train.head())
```

### Подбор гиперпараметров моделей
Решение задачи реализации модели по числовым признакам будет обеспечиватьсяс помощью следующих типов моделей: Линейная регрессия, Дерево решений, Случайный лес. Выбор набора моделей осуществлялся на основании диаграммы, представленной на рисунке 4.

С помощью функции GridSearchCV() пакета sklearn подберём оптимальный набор гиперпараметров моделей каждого из двух целевых признаков rougher.output.recovery и final.output.recovery.

*Примечание 1: в некоторых моделях значения гиперпараметров допускают достаточно большие числовые интервалы (от 1 до 101 и более); в целях сократить время обработки кода блокнота на персональном компьютере я принял решение применять интервалы от 1 до 8.*

*Примечание 2: также в функции GridSearchCV() используется аргумент scoring=smape_score для использования его в оценке; в результате этого оценочные значения отрицательные; без использования этого аргумента, оценочные значения положительные; правила интерпретации отрицательных значений в руководстве к функции GridSearchCV() не представлены, поэтому я оставлю эти значения без комментариев, достаточно будет получить набор оптимальных гиперпараметров.*

*Примечание 3: Подбор гиперпараметров модели Случайный лес работает очень долго, несмотря на сокращенные интервалы значений гиперпараметров; при этом процессор моего ноутбука загружен на 100% и все вычисления выполняются в памяти компьютера без постоянных обращений к диску. #### Подбор оптимальных параметров моделей для признака rougher.output.recovery

Подбор гиперпараметров модели Линейной регрессии сводится к определению характеристики числового ряда, т.е. признака. Если признак стандартизован, то применяется значение False, если не стандартизован, - то True.

```python
%%time
# признак rougher.output.recovery
cv = KFold(n_splits=3)
model_linear_rough = LinearRegression()
    
# словарь подбираемых значений гиперпараметров модели
hyperparameters = {'fit_intercept': [True]}

# генерация оптимальных гиперпараметров
grid_linear_rough = GridSearchCV(model_linear_rough,
                                 hyperparameters,
                                 cv=cv,
                                 scoring=smape_score,
                                 n_jobs=-1)
grid_linear_rough.fit(features_rough_train, target_rough_train)

# результаты
print('Подбор оптимальной модели Линейной регрессии')
print('Модель:', grid_linear_rough.best_estimator_)
print('Гиперпараметры:', grid_linear_rough.best_params_)
print('Оценка:', -grid_linear_rough.best_score_)
```

### Обучение оптимальных моделей
Для определения лучшей модели воспользуемся формулой расчёта итогового значени Симметричной Средней Абсолютной Процентной Ошибки.

<p><center><b>Формула 3: Итоговая Симметричная Средняя Абсолютная Процентная Ошибка (sMAPE)</b></center></p>

$$
sMAPE(итоговая) = 25\%*sMAPE(rougher) + 75\%*sMAPE(final)
$$

#### Определение лучшей модели с помощью Итоговой Симметричной Средней Абсолютной Процентной Ошибки (sMAPE)

```python
print('Итоговая sMAPE для оптимальных моделей')
print('Линейная регрессия:', (-grid_linear_rough.best_score_ * 0.25) + (-grid_linear_fine.best_score_ * 0.75))
print('Дерево решений:', (-grid_tree_rough.best_score_ * 0.25) + (-grid_tree_fine.best_score_ * 0.75))
print('Случайный лес:', (-grid_forest_rough.best_score_ * 0.25) + (-grid_forest_fine.best_score_ * 0.75))
```

С помощью метрики взвешанная sMAPE определена лучшая модель машинного обучения - Случайный лес с результатом Итоговой sMAPE равным ~8.6018. Обучим эту модель и проверим на тестовой выборке.

```python
# экземпляр оптимальной модели
model_forest_fine = RandomForestRegressor(random_state=state,
                                          max_depth=5,
                                          max_features=3,
                                          min_samples_leaf=5,
                                          n_estimators=5)  

# обучение модели
print('Обученная модель с признаком final.output.recovery:\n',
      model_forest_fine.fit(features_fine_train, target_fine_train))
```

### Тестирование оптимальных моделей

```python
# тестирование
print('Тестирование оптимальной модели')

# модель Случайный лес
# признак rougher.output.recovery
model_forest_rough_prediction = model_forest_rough.predict(features_rough_test)
model_forest_rough_smape = smape(target_rough_test, model_forest_rough_prediction)

# признак final.output.recovery
model_forest_fine_prediction = model_forest_fine.predict(features_fine_test)
model_forest_fine_smape = smape(target_fine_test, model_forest_fine_prediction)

# результат
print('Случайный лес, итоговая sMAPE:', 0.25 * model_forest_rough_smape + 0.75 * model_forest_fine_smape)
```

Следовательно, можно обоснованно резюмировать, что выбранная модель типа Случайный лес с набором оптимальных параметров является лучшей. Перейдём к выводам.
